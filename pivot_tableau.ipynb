```python
import pandas as pd
```

```python
partidas = pd.read_csv('events_clean.csv')
partidas.head()
```

```python
pivot_partidas = pd.DataFrame(columns=['time','x','y','team'])
```

```python
for ii in range(len(partidas)):
    if ii%100 == 0:
        print(ii)
    pivot_partidas.loc[len(pivot_partidas)]=[partidas.iloc[ii][0],partidas.iloc[ii][1],partidas.iloc[ii][2],'ball']
    for jj in range(0,11):
        pivot_partidas.loc[len(pivot_partidas)]=[partidas.iloc[ii][0],partidas.iloc[ii][3+2*jj],partidas.iloc[ii][4+2*jj],'team1']
        pivot_partidas.loc[len(pivot_partidas)]=[partidas.iloc[ii][0],partidas.iloc[ii][25+2*jj],partidas.iloc[ii][26+2*jj],'team2']
```

```python
pivot_partidas.tail()
```

```python
pivot_partidas.to_csv('pivot_events.csv',index=False)
```
